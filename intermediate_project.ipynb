## Вычисление среднего количества покупок в неделю для каждого пользователя по месяцам
### Загрузка данных

```python
import pandas as pd

customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
```

### Фильтрация заказов, которые считаем покупками

```python
valid_statuses = ['shipped', 'delivered']
valid_orders = orders[(orders['order_status'].isin(valid_statuses)) & 
                      (orders['order_approved_at'].notnull())]
valid_orders
```

### Объединение с таблицей клиентов для получения customer_unique_id

```python
merged_data = valid_orders.merge(customers, on='customer_id')
merged_data
```

### Группировка по уникальным идентификаторам пользователей и подсчет количества заказов

```python
customer_order_counts = merged_data.groupby('customer_unique_id')['order_id'].nunique()
customer_order_counts
```

### Определение пользователей, совершивших покупку только один раз

```python
single_purchase_customers = customer_order_counts[customer_order_counts == 1].count()
single_purchase_customers
```

### Фильтрация заказов с статусами 'canceled' и 'unavailable'

```python
undelivered_orders = orders[orders['order_status'].isin(['canceled', 'unavailable'])]
undelivered_orders
```

### Преобразование столбца order_purchase_timestamp в формат даты

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_purchase_timestamp']
```

### Извлечение года и месяца из order_purchase_timestamp

```python
undelivered_orders['purchase_year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
undelivered_orders['purchase_year_month']
```

### Группировка по месяцам и статусам

```python
undelivered_orders_grouped = undelivered_orders.groupby(['purchase_year_month', 'order_status']).size().reset_index(name='count')
undelivered_orders_grouped
```

### Вычисление среднего количества недоставленных заказов по месяцам для каждого статуса

```python
average_undelivered_per_month = undelivered_orders_grouped.groupby('order_status')['count'].mean().reset_index()
average_undelivered_per_month
```

### Объединение таблицы заказов с таблицей товарных позиций

```python
merged_orders_items = order_items.merge(orders, on='order_id')
merged_orders_items
```

### Преобразование столбца order_purchase_timestamp в формат даты

```python
merged_orders_items['order_purchase_timestamp'] = pd.to_datetime(merged_orders_items['order_purchase_timestamp'])
merged_orders_items['order_purchase_timestamp']
```

### Извлечение дня недели из временной метки покупки

```python
merged_orders_items['purchase_day_of_week'] = merged_orders_items['order_purchase_timestamp'].dt.day_name()
merged_orders_items['purchase_day_of_week']
```

### Подсчет количества покупок для каждого товара по дням недели

```python
product_day_counts = merged_orders_items.groupby(['product_id', 'purchase_day_of_week']).size().reset_index(name='count')
product_day_counts
```

### Определение самого популярного дня недели для каждого товара

```python
most_popular_day_per_product = product_day_counts.loc[product_day_counts.groupby('product_id')['count'].idxmax()]
most_popular_day_per_product
```

### Сортировка по product_id для удобства чтения

```python
most_popular_day_per_product = most_popular_day_per_product.sort_values(by='product_id').reset_index(drop=True)
most_popular_day_per_product.head(10)
```

### Объединение таблицы заказов с таблицей клиентов

```python
import numpy as np

merged_orders_customers = orders.merge(customers, on='customer_id')
merged_orders_customers
```

### Преобразование столбца order_purchase_timestamp в формат даты

```python
merged_orders_customers['order_purchase_timestamp'] = pd.to_datetime(merged_orders_customers['order_purchase_timestamp'])
merged_orders_customers.dtypes
```

### Извлечение месяца и года из временной метки покупки

```python
merged_orders_customers['purchase_month'] = merged_orders_customers['order_purchase_timestamp'].dt.to_period('M')
merged_orders_customers['purchase_month']
```

### Подсчет количества покупок для каждого пользователя по месяцам

```python
monthly_purchases = merged_orders_customers.groupby(['customer_unique_id', 'purchase_month']).size().reset_index(name='purchase_count')
monthly_purchases
```

### Функция для вычисления количества недель в месяце

```python
def weeks_in_month(period):
    start_date = period.start_time
    end_date = period.end_time
    days_in_month = (end_date - start_date).days + 1
    weeks = days_in_month / 7
    return weeks
```

### Добавление столбца с количеством недель в месяце

```python
monthly_purchases['weeks_in_month'] = monthly_purchases['purchase_month'].apply(weeks_in_month)
monthly_purchases
```

### Вычисление среднего количества покупок в неделю для каждого пользователя по месяцам

```python
monthly_purchases['avg_purchases_per_week'] = monthly_purchases['purchase_count'] / monthly_purchases['weeks_in_month']
monthly_purchases.head(10)
```

## Когортный анализ
### Загрузка данных

```python
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
```

```python
customers
```

```python
order_items
```

### Фильтрация заказов по статусам 'shipped' и 'delivered' и наличию даты подтверждения оплаты

```python
filtered_orders = orders[(orders['order_status'].isin(['shipped', 'delivered'])) & (orders['order_approved_at'].notnull())]
filtered_orders
```

### Преобразование столбцов с датами к типу datetime

```python
filtered_orders['order_purchase_timestamp'] = pd.to_datetime(filtered_orders['order_purchase_timestamp'])
filtered_orders['order_approved_at'] = pd.to_datetime(filtered_orders['order_approved_at'])
filtered_orders['order_delivered_carrier_date'] = pd.to_datetime(filtered_orders['order_delivered_carrier_date'])
filtered_orders['order_delivered_customer_date'] = pd.to_datetime(filtered_orders['order_delivered_customer_date'])
filtered_orders['order_estimated_delivery_date'] = pd.to_datetime(filtered_orders['order_estimated_delivery_date'])
```

### Соединение таблиц filtered_orders и customers

```python
merged_data = pd.merge(filtered_orders, customers, on='customer_id')
merged_data
```

### Определение первой покупки каждого пользователя

```python
first_order = merged_data.groupby('customer_unique_id').order_purchase_timestamp.min().reset_index()
first_order.columns = ['customer_unique_id', 'first_order_date']
first_order
```

### Добавление столбца с датой первой покупки к заказам

```python
merged_data = pd.merge(merged_data, first_order, on='customer_unique_id')
merged_data
```

### Определение месяца первой покупки

```python
merged_data['first_order_month'] = merged_data['first_order_date'].dt.to_period('M')
merged_data
```

### Определение месяца текущего заказа

```python
merged_data['order_month'] = merged_data['order_purchase_timestamp'].dt.to_period('M')
merged_data
```

### Создание таблицы для когортного анализа

```python
cohorts = merged_data.groupby(['first_order_month', 'order_month']).agg({'customer_unique_id': 'nunique'}).reset_index()
cohorts.rename(columns={'customer_unique_id': 'num_users'}, inplace=True)
cohorts
```

### Переименуем периоды для удобства

```python
cohorts['cohort_period'] = (cohorts.order_month - cohorts.first_order_month).apply(lambda x: x.n)
cohorts
```

### Поворот таблицы для удобства вычислений

```python
cohort_pivot = cohorts.pivot_table(index='first_order_month', columns='cohort_period', values='num_users')
cohort_pivot
```

### Расчет ретеншна

```python
cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
```

```python
cohort_size
```

```python
retention_matrix
```

### Определение когорты с самым высоким retention на 3-й месяц

```python
max_retention_3_month = retention_matrix[3].idxmax()
max_retention_value = retention_matrix[3].max()
```

```python
max_retention_3_month, max_retention_value
```

### Результаты показывают, что когорта пользователей, совершивших свою первую покупку в марте 2017 года, имеет самый высокий retention на 3-й месяц, составляющий примерно 0.00393 (или 0.393%)

## RFM-сегментирование
### Загрузка данных

```python
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
```

```python
customers
```

```python
orders
```

```python
order_items
```

### Объединение таблиц

```python
merged_df = pd.merge(orders, order_items, on='order_id')
merged_df = pd.merge(merged_df, customers, on='customer_id')
merged_df
```

### Фильтрация данных

```python
filtered_df = merged_df[
    (merged_df['order_status'].isin(['shipped', 'delivered'])) &
    (merged_df['order_approved_at'].notna())
]
filtered_df
```

### Преобразование столбцов дат в формат datetime

```python
filtered_df['order_purchase_timestamp'] = pd.to_datetime(filtered_df['order_purchase_timestamp'])
filtered_df['order_approved_at'] = pd.to_datetime(filtered_df['order_approved_at'])
filtered_df['order_delivered_customer_date'] = pd.to_datetime(filtered_df['order_delivered_customer_date'])
```

```python
filtered_df.head()
```

### Установим текущую дату для расчета recency

```python
from datetime import datetime

current_date = datetime.now()
current_date
```

### Создаем датафрейм для метрик RFM

```python
rfm_df = filtered_df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'price': 'sum'  # Monetary
}).reset_index()
rfm_df
```

### Переименовываем колонки для удобства

```python
rfm_df.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm_df.head()
```

### Разделение метрик на квантильные сегменты

```python
rfm_df['R_Segment'] = pd.qcut(rfm_df['Recency'], 3, labels=[1, 2, 3])
rfm_df['F_Segment'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 3, labels=[1, 2, 3])
rfm_df['M_Segment'] = pd.qcut(rfm_df['Monetary'], 3, labels=[1, 2, 3])
rfm_df
```

### Объединение сегментов в один RFM сегмент

```python
rfm_df['RFM_Segment'] = rfm_df['R_Segment'].astype(str) + rfm_df['F_Segment'].astype(str) + rfm_df['M_Segment'].astype(str)
rfm_df
```

### Подсчет RFM значения (для оценки)

```python
rfm_df['RFM_Score'] = rfm_df[['R_Segment', 'F_Segment', 'M_Segment']].sum(axis=1)
rfm_df
```

### Функция для вычисления границ квантилей

```python
def calculate_quantile_bounds(df, column, quantiles=[0.33, 0.66, 1.0]):
    quantile_bounds = df[column].quantile(quantiles).tolist()
    quantile_bounds = [df[column].min()] + quantile_bounds
    return [(quantile_bounds[i], quantile_bounds[i+1]) for i in range(len(quantile_bounds)-1)]
```

### Расчет границ для каждой метрики

```python
recency_bounds = calculate_quantile_bounds(rfm_df, 'Recency')
frequency_bounds = calculate_quantile_bounds(rfm_df, 'Frequency')
monetary_bounds = calculate_quantile_bounds(rfm_df, 'Monetary')
recency_bounds, frequency_bounds, monetary_bounds
```

### Создание таблицы интерпретации сегментов

```python
interpretation_df = pd.DataFrame({
    'Segment': ['1', '2', '3'],
    'Recency_Bounds': recency_bounds,
    'Frequency_Bounds': frequency_bounds,
    'Monetary_Bounds': monetary_bounds
})
interpretation_df
```

### Функция для получения описания сегмента

```python
def get_segment_description(segment_code, recency_bounds, frequency_bounds, monetary_bounds):
    r = int(segment_code[0])
    f = int(segment_code[1])
    m = int(segment_code[2])
    return {
        'RFM_Segment': segment_code,
        'Recency_Bounds': recency_bounds[r-1],
        'Frequency_Bounds': frequency_bounds[f-1],
        'Monetary_Bounds': monetary_bounds[m-1]
    }
```

### Генерация описания для всех возможных RFM-сегментов

```python
segment_descriptions = []
for r in range(1, 4):
    for f in range(1, 4):
        for m in range(1, 4):
            segment_code = f"{r}{f}{m}"
            description = get_segment_description(segment_code, recency_bounds, frequency_bounds, monetary_bounds)
            segment_descriptions.append(description)
```

### Преобразование в датафрейм для удобства отображения

```python
segment_descriptions_df = pd.DataFrame(segment_descriptions)
segment_descriptions_df
```
